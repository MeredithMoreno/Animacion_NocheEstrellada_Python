```python


def recortar_escalar_imagen(ruta_imagen, ancho_recorte, alto_recorte):
    # Abre la imagen original
    imagen = Image.open(ruta_imagen)

    # Obtiene las dimensiones de la imagen original
    ancho_original, alto_original = imagen.size

    # Calcula la relacin de aspecto de la imagen original
    relacion_aspecto = ancho_original / alto_original

    # Calcula las nuevas dimensiones manteniendo la relacin de aspecto
    if ancho_recorte / alto_recorte > relacion_aspecto:
        # Si el recorte es ms ancho que la relacin de aspecto de la imagen original
        nuevo_ancho = int(alto_recorte * relacion_aspecto)
        nuevo_alto = alto_recorte
    else:
        # Si el recorte es ms alto que la relacin de aspecto de la imagen original
        nuevo_ancho = ancho_recorte
        nuevo_alto = int(ancho_recorte / relacion_aspecto)

    # Recorta la imagen al centro
    izquierda = (ancho_original - nuevo_ancho) / 2
    arriba = (alto_original - nuevo_alto) / 2
    derecha = (ancho_original + nuevo_ancho) / 2
    abajo = (alto_original + nuevo_alto) / 2
    imagen = imagen.crop((izquierda, arriba, derecha, abajo))

    # Escala la imagen al tamao del recorte
    imagen = imagen.resize((ancho_recorte, alto_recorte), Image.ANTIALIAS)

    return imagen

imagen_escalada = recortar_escalar_imagen('lanoche.png', 'lanoche2.png', 'lanoche2.png')
```
